```python
import pandas as pd
import plotly.graph_objects as go
import plotly.figure_factory as ff
import plotly.express as px
import json
import os, glob
import numpy as np
from scipy.stats import shapiro, stats, spearmanr
import seaborn as sns


from IPython.display import HTML
import random

def hide_toggle(for_next=False):
    this_cell = """$('div.cell.code_cell.rendered.selected')"""
    next_cell = this_cell + '.next()'

    toggle_text = 'Show/hide'  # text shown on toggle link
    target_cell = this_cell  # target cell to control with toggle
    js_hide_current = ''  # bit of JS to permanently hide code in current cell (only when toggling next cell)

    if for_next:
        target_cell = next_cell
        toggle_text += ' next cell'
        js_hide_current = this_cell + '.find("div.input").hide();'

    js_f_name = 'code_toggle_{}'.format(str(random.randint(1,2**64)))

    html = """
        <script>
            function {f_name}() {{
                {cell_selector}.find('div.input').toggle();
            }}

            {js_hide_current}
        </script>

        <a href="javascript:{f_name}()">{toggle_text}</a>
    """.format(
        f_name=js_f_name,
        cell_selector=target_cell,
        js_hide_current=js_hide_current, 
        toggle_text=toggle_text
    )

    return HTML(html)


df = pd.read_csv("2021.csv")
df.drop(['Explained by: Log GDP per capita', 'Explained by: Social support',
        'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices',
        'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual'], axis=1, inplace=True)

df_pop = pd.read_csv("population.csv")
hide_toggle()
```

### 3. Wczytanie danych ze strony "Population"
Poza wczytaniem danych, uzupełniam brakujące państwa o liczbę ludności w 2020 roku.

```python
df_pop = df_pop[['Country (or dependency)', 'Population (2020)']]
df_pop_dict = dict(zip(df_pop['Country (or dependency)'], df_pop['Population (2020)']))
df_pop_dict['Czech Republic'] = df_pop_dict['Czech Republic (Czechia)']
df_pop_dict['Taiwan Province of China'] = df_pop_dict['Taiwan']
df_pop_dict['Kosovo'] = 1873000
df_pop_dict['North Cyprus'] = 1207359
df_pop_dict['Hong Kong S.A.R. of China'] = df_pop_dict['Hong Kong']
df_pop_dict['Congo (Brazzaville)'] = df_pop_dict['Congo']
df_pop_dict['Palestinian Territories'] = df_pop_dict['State of Palestine']
df_pop_dict['Swaziland'] = 1160000
df_pop_dict['Ivory Coast'] = 26378274 
```

Żeby możliwa była lokalizacja państw z użyciem nazw znajdujących się w pliku o nazwie custom.geo.json, dostosowuję nazwy we wczytanej uprzednio tablicy z danymi o szczęściu, używając do tego ręcznie skonfigurowanego słownika konwersji nazw. Następnie do danych o szczęściu dopisuję kolumnę z informacjami o populacji.

```python
world_path = 'custom.geo.json'
with open(world_path) as f:
    geo_world = json.load(f)

population_col = []
for country1 in df['Country name']:
    if country1 in df_pop_dict:
        population_col.append(df_pop_dict[country1])
    else:
        print("ERROR", country1)
   
df['Population'] = population_col    
    
country_conversion_dict = {
    'Dominican Rep.': 'Dominican Republic',
    'Cyprus U.N. Buffer Zone': 'Cyprus',
    'N. Cyprus': 'North Cyprus',
    'Hong Kong': 'Hong Kong S.A.R. of China',
    'Lao PDR': 'Laos',
    'Palestine': 'Palestinian Territories',
    'Taiwan': 'Taiwan Province of China',
    'Guinea-Bissau': 'Guinea',
    'Bosnia and Herz.': 'Bosnia and Herzegovina',
    'Czech Rep.': 'Czech Republic'
}    
    
found = []
missing = []
countries_geo = []
tmp = df.set_index('Country name')

for country in geo_world['features']:
    
    country_name = country['properties']['name']
    country_name = country_conversion_dict[country_name] if country_name in country_conversion_dict.keys() else country_name
    
    if country_name in tmp.index:
        found.append(country_name)
        geometry = country['geometry']
        
        countries_geo.append({
            'type': 'Feature',
            'geometry': geometry,
            'id': country_name
        })
    
    else:
        missing.append(country_name)
        
        
print(f'Countries found    : {len(found)}')
print(f'Countries not found: {len(missing)}')
geo_world_ok = {'type': 'FeatureCollection', 'features': countries_geo}  
#print(missing)
```

### a) Ladder score

```python
ladder_score = df['Ladder score']
n = len(ladder_score)
maxi =  max(ladder_score)
mini =  min(ladder_score)
rozstep = round(maxi - mini, 3)
print("Maksimum: " , maxi, "Minimum: " , mini, "Rozstęp: ", rozstep)

srednia = round(ladder_score.mean(), 3)
mediana = round(ladder_score.median(), 3)
variance = round(ladder_score.var(), 3)
odchylenie = round(variance**(1/2), 3)
skosnosc = round(ladder_score.skew(), 3)
kurtoza = round(ladder_score.kurt(), 3)
Q3 = round(ladder_score.quantile(q=0.75), 3)
Q1 = round(ladder_score.quantile(q=0.25), 3)
IQR = round(Q3 - Q1, 3)

print("Średnia: ", srednia, " Mediana: ", mediana, " Wariancja: ", variance, " Odchylenie: ", odchylenie)
print("Skośność: ", skosnosc, " Kurtoza: ", kurtoza)
print("Pierwszy kwartyl: ", Q1, " Trzeci kwartyl: ", Q3, " Rozstęp międzykwartylowy: ", IQR)
```

Warto zaznaczyć, że średnia dla Świata, jest policzona ze średnich krajów, które się w niej znajdują. Jeżeli liczności prób danych krajów różniłyby się między sobą drastycznie, tak obliczona średnia mogłaby się różnić od tej policzonej, dla Świata, gdyby zsumowało się wartości 'Ladder score' wszystkich osób i podzieliło przez ich liczbę.


Średnia jest bardzo zbliżona do mediany, co świadczy o tym, że jest mało obserwacji odstających. Może to również wynikać z tego, że obliczona powyżej średnia jest policzona ze średnich zaobserwowanych we wszystkich krajach. Pierwszy kwartyl wynosi 4.852, co oznacza, że 25% obserwacji jest mniejszych lub równych od tej wartości. Trzeci kwartyl wynosi 6.255, a zatem 75% obserwacji osiąga wartość mniejszą lub równą 6.255.

Liczbę przedziałów w histogramie dotyczącym rozkładu 'Ladder score', czyli poziomu szczęścia, obliczyłam z użyciem następującego wzoru:



k - liczba przedziałów   
x_max, x_min - maksymalna, minimalna wartość w próbie   
n - liczba obserwacji w próbie   
IQR - rozstęp międzykwartylowy  

```python
liczba_przedzialow = round(((rozstep)/(IQR* 2.64 * n**(-1/3))), 0)
szerokosc_przedzialu = rozstep / liczba_przedzialow
```

Należy również zaznaczyć, że minimum dla kolumny 'Ladder score' wyniosło  2.523, a początek pierwszego przedziału zaczyna się od wartości 2.521. Jest tak, ponieważ przedziały są lewostronnie otwarte i należało wydłużyć ten przedział, aby wartość minimalna była zawarta w histogramie.


Rozkład jest bimodalny. Pierwszy pik występuje w okolicach wartości 4.8, a drugi pik w okolicy wartości 6.2.

### Test Shapiro-Wilka


H0: Rozkład zmiennej ladder_score jest zbliżony do rozkładu normalnego.  
H1: Rozkład zmiennej ladder_score nie jest zbliżony do rozkładu normalnego. 

Założony poziom istotności to alfa=0.05.

```python
stat, p = shapiro(ladder_score)
print('Wartość statystyki=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
	print('Nie odrzucamy H0')
else:
	print('Odrzucamy H0')
```

Wniosek:  Rozkład zmiennej ladder_score jest zbliżony do rozkładu normalnego.

### c)  Wykresy rozproszenia oraz wyznaczenie współczynników korelacji Spearmana

Przeprowadzenie testu Pearsona na istotność korelacji wymaga między innymi, aby obie testowane zmienne miały rozkład normalny. Z tego powodu do zbadania istotności korelacji pomiędzy zmiennymi  użyję testu Spearmana, który nie wymaga normalności rozkładu próbek (ponieważ jest to test nieparametryczny), ale zakłada między innymi, że:  
1.  Próba jest losowa.    
2.  Występuje zależność monotoniczna pomiędzy zmiennymi (co można odczytać z wykresu).  
3. Zmienne są przynajmniej typu porządkowego.  
4. Dane są sparowane.  
5. Obserwacje są niezależne.  

Wszystkie te warunki są spełnione dla par zmiennych ladder_score oraz odpowiednio: 
logged_gdp_per_capita, social_support, life_exp, freedom_life_choices, generosity, corruption.

##### Ladder score & Logged GDP per capita

```python
sns.lmplot(x='Ladder score', y='Logged GDP per capita', data=df, fit_reg=False)
```

Z wykresu można odczytać, że wzrostowi wyników Ladder score towarzyszy wzrost wyników Logged GDP per capita. Można podejrzewać, że istnieje korelacja pomiędzy tymi zmiennymi. Wykonajmy test istotności korelacji rangowej Spearmana. Jest to test dwustronny.

H0: Próbki są nieskorelowane.  
H1: Próbki są skorelowane.  

Zakładamy poziom istotności na alfa=0.01

```python
coef_logged_gdp_per_capita, p = spearmanr(ladder_score, logged_gdp_per_capita)
print('Spearmans correlation coefficient: %.3f' % coef_logged_gdp_per_capita)
print('P-wartość: %.3f' % p)

alpha = 0.01
if p > alpha:
	print('Nie odrzucamy H0')
else:
	print('Odrzucamy H0')
```

Mamy podstawy do tego, by odrzucić H0. Próbki ladder_score oraz logged_gdp_per_capita są skorelowane. Jest to silna korelacja.

### 3. Test dwóch zbiorowości na równość średnich

* H0: Średnia szczęścia w Europie Zachodniej jest równa średniej szczęścia w Europie Centralnej i Wschodniej 
* H1: Średnia szczęścia w Europie Zachodniej nie jest równa średniej szczęścia w Europie Centralnej i Wschodniej 

#### Weryfikacja wymagań
Sprawdzam, czy rozkład 'Ladder score' dla Europy Zachodniej jest normalny. Używam do tego testu Shapiro-Wilka. Te same kroki powtarzam dla Europy Centralno-Zachodniej. Założony poziom istotności to alfa=0.05.

##### Europa Zachodnia
Test Shapiro-Wilka:

* H0: Rozkład zmiennej europa_zachodnia jest zbliżony do normalnego.
* H1: Rozkład zmiennej europa_zachodnia nie jest zbliżony do normalnego.

```python
europa_zachodnia = df.query(" `Regional indicator` == 'Western Europe' ")['Ladder score']

print("Liczba danych w europa_zachodnia: ", len(europa_zachodnia))

stat, p = shapiro(europa_zachodnia)
print('Wartość statystyki=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
	print('Nie odrzucamy H0')
else:
	print('Odrzucamy H0')
```

Skoro nie udało się odrzucić H0 na poziomie istotności alfa=0.05, to przyjmujemy, że rozkład europa_zachodnia jest normalny.

##### Europa Centralno-Wschodnia

Test Shapiro-Wilka:

* H0: Rozkład zmiennej europa_centrwsch jest zbliżony do normalnego.
* H1: Rozkład zmiennej europa_centrwsch nie jest zbliżony do normalnego.

```python
europa_centrwsch = df.query(" `Regional indicator` == 'Central and Eastern Europe' ")['Ladder score']

print("Liczba danych w europa_centrwsch: ", len(europa_centrwsch))

stat, p = shapiro(europa_centrwsch)
print('Wartość statystyki=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
	print('Nie odrzucamy H0')
else:
	print('Odrzucamy H0')
```

Skoro nie udało się odrzucić H0 na poziomie istotności alfa=0.05, to przyjmujemy, że rozkład europa_centrwsch jest normalny.

##### Przeprowadzenie testu

Populacje te są od siebie niezależne. Liczba danych w obu próbach jest mniejsza od 30 (odpowiednio 21 dla Europy Zachodniej oraz 17 dla Europy Centralno-Wschodniej), dlatego należy zastosować test T niesparowany.
Test ten zakłada równość wariancji w obu populacjach.

H0: Średnia szczęścia w Europie Zachodniej jest równa średniej szczęścia w Europie Centralno-Wschodniej.   
H1: Średnia szczęścia w Europie Zachodniej nie jest równa średniej szczęścia w Europie Centralno-Wschodniej. 

Założony poziom istotności to alfa=0.05.

```python
stat, p = stats.ttest_ind(europa_centrwsch, europa_zachodnia, equal_var=True)
print('Wartość statystyki=%.3f, p=%.3f' % (stat, p))

alpha = 0.05
if p > alpha:
	print('Nie odrzucamy H0')
else:
	print('Odrzucamy H0')
```

##### Wniosek
Średnie szczęścia w Europie Zachodniej i Europie Centralno-Wschodniej są różne.
